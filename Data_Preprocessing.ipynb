```python
# Basic imports
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel('macro_economic.xlsx')
data.head(5)
```

```python
data.shape
```

```python
train = pd.read_csv('train.csv')
train.head()
```

```python
train.shape
```

```python
test = pd.read_csv('test.csv')
test.head()
```

```python
# Appending train and test data one over the other
df = train.append(test).reset_index(drop=True)
df.head(10)
```

```python
# Removing 2016 data from econ_data as train and test doesn't have 2016 data
econ_2015 = data[:84]
econ_2015.head(10)
```

```python
# copy each row in econ_data thrice so as to match (train + test) data
econ_data_copy = pd.DataFrame(np.repeat(econ_2015.values,3,axis=0))
econ_data_copy.head(4)
```

### Null values check on econ_data:

```python
# Checking dataset if there are any null values present
econ_data_copy.isnull().values.any()
```

But the econ_data has columns with special values like '?', which has to be addressed.

```python
# Concatenating (train + test data modified) and econ_data
df_final = pd.concat([df, econ_data_copy], axis=1)
df_final.head(10)
```

```python
# Adding appropriate column names to df_final
df_final.columns = list(df.columns) + list(econ_2015.columns)
```

```python
# The final dataset to work with
df_final.T
```

---
## Dropping Unnecessary Features:

```python
# 1. Removing redundant 'Year-Month'  and   PartyInPower
df_final = df_final.drop('Year-Month', axis=1)

df_final['PartyInPower'].value_counts()
```

#### party power has only democrats as value

```python
# 3. how many Advertising Expenses

econ_2015['AdvertisingExpenses (in Thousand Dollars)'][econ_2015['AdvertisingExpenses (in Thousand Dollars)'] == '?'].count()
```

```python
df_final = df_final.drop(['PartyInPower', 'AdvertisingExpenses (in Thousand Dollars)'], axis = 1)
```

#### Since Advertising Expenses is '?' for 73 of the possible 84 values; we can safely eliminate this feature for now. 

---

## Exploratory Data Analysis on df_final:

```python
df_final.info()
```

```python
cols = list(df_final.columns)
cols.remove('ProductCategory')
cols.remove('Sales(In ThousandDollars)')
cols.remove('Year')
cols.remove('Month')
cols
```

```python
for col in cols:
    df_final[col] = df_final[col].astype('float64')
```

```python
df_final.info()
```

```python
df_final.describe(include='all')
```

```python
df_final.corr()
```

### One-Hot Encoding ProductCategory

```python
column = ['ProductCategory']

df_final = pd.get_dummies(data=df_final, prefix=None, columns=column, dtype='int', drop_first=True)
```

```python
df_final.T
```

## Splitting df_final to train_final and test_final:

```python
test_final = df_final[df_final['Year'] == 2015]
test_final.head(10)
```

```python
# Resetting index of test_final
test_final = test_final.reset_index(drop=True)
test_final = test_final.drop('Sales(In ThousandDollars)', axis=1)
```

```python
test_final.info()
```

```python
train_final = df_final[:216]
train_final.head()
```

```python
# Imputing missing values in sales in train data with median
train_final = train_final.fillna(train_final.median())
train_final.head()
```

```python
train_final.to_csv('train_final.csv')
```
